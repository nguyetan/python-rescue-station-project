```python
# Chọn cột và sắp xếp lại thứ tự cột
df_selected = data[['Id', 'FromPoint', 'ToPoint', 'chieudai']]

# Hiển thị kết quả
print(df_selected)
```

```python
# Nhập id x và id y từ người dùng
x = int(input("Nhập id x: "))
y = int(input("Nhập id y: "))

# Lọc ra các dòng có Id nằm trong khoảng từ id x đến id y
filtered_df = df_selected[(df_selected['Id'] >= x) & (df_selected['Id'] <= y)]

# In ra DataFrame mới
print(filtered_df)
```

```python
import networkx as nx

# Tạo đồ thị vô hướng từ dữ liệu filtered_df
G = nx.Graph()

# Thêm các cạnh vào đồ thị với trọng số tương ứng là 'chieudai'
for index, row in filtered_df.iterrows():
    G.add_edge(row['FromPoint'], row['ToPoint'], weight=row['chieudai'])

# Hàm tính khoảng cách network giữa các cặp đỉnh
def calculate_network_distances(graph):
    distances = dict(nx.all_pairs_dijkstra_path_length(graph, weight='weight'))
    return distances

# Tính toán khoảng cách
network_distances = calculate_network_distances(G)

# Chuyển đổi kết quả thành DataFrame
distance_records = []
for start_point, end_points in network_distances.items():
    for end_point, distance in end_points.items():
        distance_records.append({
            'FromPoint': start_point,
            'ToPoint': end_point,
            'chieudai': distance
        })

distance_df = pd.DataFrame(distance_records)

# Hiển thị DataFrame kết quả
print(distance_df)
```

```python
# Trộn các kết quả vào chung 3 cột chieudai, FromPoint, ToPoint
merged_df = pd.concat([distance_df, filtered_df[['FromPoint', 'ToPoint', 'chieudai']]])

# Reset index để tránh index trùng lặp
merged_df.reset_index(drop=True, inplace=True)

# Hiển thị kết quả
print(merged_df)
```

```python
# Đổi tên các cột
df = merged_df.rename(columns={"FromPoint": "StartPoint", "ToPoint": "EndPoint", "chieudai": "Distance"})

print(df)
```
